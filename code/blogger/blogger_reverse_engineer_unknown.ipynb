```python
unknown_concat = concatenate_sentences(unknown_preprocessed, length_threshold=500, threshold_type='word')
```

```python
save_as_jsonl(unknown_concat, f"{blogger_loc}unknown_combined.jsonl")
```

```python
unknown_concat
```

```python
unknown_final = unknown_concat.drop_duplicates(subset='id', keep='first')
```

```python
# Keep the specified columns
unknown_final = unknown_final[['id', 'author_id', 'gender', 'word_count', 'age', 'topic', 'sign', 'date', 'text']]

# Rename 'id' to 'doc_id'
unknown_final.rename(columns={'id': 'doc_id'}, inplace=True)
```

```python
save_as_jsonl(unknown_final, f"{blogger_loc}unknown_final.jsonl")
```

```python
metadata = unknown_final[['doc_id', 'author_id', 'topic']].copy()

metadata.rename(columns={'doc_id': 'doc_id_y', 'author_id': 'author_id_y', 'topic': 'topic_y'}, inplace=True)
```

```python
metadata
```

```python
# Add the sample_id column with the row number (starting from 1)
metadata['sample_id'] = range(1, len(metadata) + 1)

# Move the sample_id column to the front
cols = ['sample_id'] + [col for col in metadata.columns if col != 'sample_id']
metadata = metadata[cols]
```
