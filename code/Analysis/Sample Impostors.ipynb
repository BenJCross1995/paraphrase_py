```python
import random
import pandas as pd

from pathlib import Path
from read_and_write_docs import read_jsonl
```

```python
def random_jsonl(directory: str | Path) -> Path | None:
    """
    Return a random .jsonl file Path from *directory*.
    If there are no .jsonl files, return None.
    
    Parameters
    ----------
    directory : str | pathlib.Path
        Folder in which to search for .jsonl files.
    
    Returns
    -------
    pathlib.Path | None
        Path object pointing to the chosen file, or None if none found.
    """
    directory = Path(directory).expanduser().resolve()
    jsonl_files = list(directory.glob("*.jsonl"))
    chosen = random.choice(jsonl_files)
    
    print(chosen.name)
    
    return chosen

def show_random_samples(df: pd.DataFrame, n: int = 10) -> None:
    """
    Pick *n* random rows and print them in the requested format.

    Expected columns
    ----------------
    parascore_free : numeric or str
    rephrased      : str
    """
    if df.empty:
        print("DataFrame is empty.")
        return

    # Take a random subset (n or all, whichever is smaller)
    subset = df.sample(n=min(n, len(df)), random_state=None)

    # Iterate and display
    for idx, (row_index, row) in enumerate(subset.iterrows(), start=1):
        print(f"Sample {idx}, Parascore {row['parascore_free']}")
        print(row['rephrased'], end="\n\n")
```
